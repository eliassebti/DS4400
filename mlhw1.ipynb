```python
import pandas as pd
import numpy as np




def genMat():
    return np.array([[np.random.randint(-10, 11),np.random.randint(-10, 11),np.random.randint(-10, 11)],[np.random.randint(-10, 11),np.random.randint(-10, 11),np.random.randint(-10, 11)],[np.random.randint(-10, 11),np.random.randint(-10, 11),np.random.randint(-10, 11)]])

A = genMat()
B = genMat()
C = genMat()

Ainv = np.linalg.inv(A)
Binv = np.linalg.inv(B)
Cinv = np.linalg.inv(C)

print(A@Ainv)
print(B@Binv)
print(C@Cinv)
# Question 4.2: You can see the inverse time the matrix equals the identity matrix.Question 4.2: You can see the inverse time the matrix equals the identity matrix.
```
